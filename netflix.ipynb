```python
import pandas as pd 
file_path = ("C:\\Users\\sairu\\Downloads\\netflix_titles.csv\\netflix_titles.csv" )
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
(df.isnull().sum() /len(df)) * 100
```

```python
df['director'].fillna('unknown',inplace = True)
df.dropna(subset=['title'],inplace = True)
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'])

# Remove leading/trailing spaces from object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

```python
df.isna().sum()
df
```

```python
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df.isna().sum()
```

```python
df.dropna(subset=['date_added'],inplace = True)
df['rating'].fillna("not rated",inplace = True)
df.isna().sum()
```

```python
df.dropna(subset=['duration'],inplace=True)
df.isna().sum()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.countplot(x='type',data=df)
plt.title('Count of movies vs Tv shows')
plt.show()
```

```python
sns.countplot(x='rating',data=df)
plt.title('Distribution of Ratings')
plt.xticks(rotation = 45)
plt.show()
```

```python
df['release_year'].hist(bins =20)
plt.title('Disturbution of Release Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

# Multi varient Analysis

```python
# multi varient analysis
sns.countplot(x='release_year', hue='type', data=df)
plt.title('Movies vs TV Shows per Release Year')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x='rating',hue = 'type',data=df)
plt.title("Ratings by Type")
plt.show()
```

# Genre analysis

```python
genre_series = df['listed_in'].str.split(', ').explode()
genre_counts = genre_series.value_counts().head(10)
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title('Top 10 Netflix Genres')
plt.show()
```

# Time analysis

```python
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind = 'bar',figsize =(12,5))
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

# Feature Engineering

```python
df['Year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['duration_minutes'] = df['duration'].str.extract('(\d+)').astype(int)
df_exploded = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')
df['is_movie'] = df['type'] == 'Movie'
df['is_tv_show'] = df['type'] == 'TV Show'
df
```

```python
# Count movies and TV shows
num_movies = df['is_movie'].sum()
num_tv_shows = df['is_tv_show'].sum()

print(f"Total Movies: {num_movies}")
print(f"Total TV Shows: {num_tv_shows}")
```

```python
# Split multiple genres
df_exploded = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')

# Top 10 genres overall
top_genres = df_exploded['listed_in'].value_counts().head(10)

sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title('Top 10 Netflix Genres')
plt.show()
```

```python
# Movies vs TV Shows genres
movies_genres = df_exploded[df_exploded['is_movie']]['listed_in'].value_counts().head(10)
tv_genres = df_exploded[df_exploded['is_tv_show']]['listed_in'].value_counts().head(10)

# Movies
sns.barplot(x=movies_genres.values, y=movies_genres.index)
plt.title('Top 10 Genres for Movies')
plt.show()

# TV Shows
sns.barplot(x=tv_genres.values, y=tv_genres.index)
plt.title('Top 10 Genres for TV Shows')
plt.show()
```

```python
# Extract numeric duration for movies
df['duration_minutes'] = df['duration'].str.extract('(\d+)').astype(float)

# Plot movie durations
sns.histplot(df[df['is_movie']]['duration_minutes'], bins=20)
plt.title('Distribution of Movie Durations')
plt.xlabel('Minutes')
plt.show()
```
